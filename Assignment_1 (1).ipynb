## 9. Generate the series of dates from 1st Jan, 2023 to 10th Feb, 2023

```python
import datetime
start=datetime.date(2023,1,1)
k=40
res=[]
for day in range(k):
  date=(start+datetime.timedelta(days=day)).isoformat()
  res.append(date)
print("Dates-"+str(res))  
```
